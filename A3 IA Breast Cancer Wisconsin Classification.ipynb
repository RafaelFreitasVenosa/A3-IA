```python
import pandas as pd
# Lendo o arquivo data.csv e atribuindo seus dados para o dataframe df
df = pd.read_csv("data.csv")
```

```python
# Exibindo as colunas
df.columns
df.drop('Unnamed: 32', axis= 1, inplace= True)

# Definimos X (variáveis preditoras) e y (variável target).
x = df[['id', 'radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]

y = df['diagnosis']

#df.drop('diagnosis', axis= 1, inplace= True)
print(x)
print(y)
```

```python
from sklearn import preprocessing

#criando labelEncoder
le = preprocessing.LabelEncoder()

alvo = le.fit_transform(y)
print(alvo)

y = alvo
```

```python
#Método 1 - Naive Bayes (ENTREGA 04)

from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split

# treino e teste
(x_train, x_test, y_train, y_test) = train_test_split(x,y)

#Cria um classificador Gaussiano
modelo = GaussianNB()

# treinando o modelo utilizando o conjunto de treino
modelo.fit(x_train,y_train)

# validando o modelo utilizando o conjunto de teste
precisao = str(round(modelo.score(x_test,y_test) * 100, 2))+"%"

# imprimindo o resultado
print("A acurácia do modelo 1 foi", precisao)
```

```python
#Método 2 - KNN (ENTREGA 05)

import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split

# Dividimos a base de dados em grupos de teste e de treinamento
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)

# Modelo que representa o método KNN de Regressão, utilizando k = 3
model = KNeighborsClassifier(n_neighbors=3)

# Aplicamos os dados de treinamento ao modelo
model.fit(x_train,y_train)

# Cálculo da precisão com formatação 99% ao invés de 0.99
precisaoKnn = str(round(model.score(x_test,y_test) * 100, 2))+"%"

print("A acurácia do modelo 2 foi", precisaoKnn)

```

```python
#Método 3 - Redes Neurais (ENTREGA 06)

from sklearn.neural_network import MLPClassifier
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

# Dividimos a base de dados em grupos de teste e de treinamento
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state =1)

# Modelo que representa o método Redes Neurais
clf = MLPClassifier(solver='lbfgs', activation='logistic', hidden_layer_sizes=(50, 20, 30))

# Aplicamos os dados de treinamento ao modelo
clf.fit(x_train, y_train)

# Cálculo da precisão
precisaoRedesNeurais = str(round(clf.score(x_test,y_test) * 100, 2))+"%"

print("A acurácia do modelo 3 foi", precisaoRedesNeurais)
```

```python
#Método 4 - Àvore de Decisão

from sklearn import tree
from sklearn.model_selection import train_test_split
import pandas as pd
import graphviz

#Divide a base de dados em treinamento e teste
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
#Cria um modelo da arvore de decisão e atribui para o objeto clf
clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=3)
#Executa o treinamento
clf = clf.fit(x_train, y_train)
#Aplica o método para a previsão
predictions = clf.predict(x_test)
# Cálculo da precisão
precisao = str(round(clf.score(x_test,y_test) * 100, 2))+"%"
print("A precisão do método foi de: " + precisao)

```
